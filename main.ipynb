```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('train.txt',sep=';',names=['text','emotion'], header=None)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
unique_emotions=df['emotion'].unique()
```

```python
emotion_numbers={}
i=0
for emo in unique_emotions:
    emotion_numbers[emo]=i
    i+=1
df['emotion']=df['emotion'].map(emotion_numbers)
```

```python
emotion_numbers
```

```python
df['text']=df['text'].str.lower()
```

```python
import string 
def remove_punctuation(text):
    return text.translate(str.maketrans('','',string.punctuation))
df['text']=df['text'].apply(remove_punctuation)
```

```python
def remove_numbers(text):
    return text.translate(str.maketrans('','',string.digits))
df['text']=df['text'].apply(remove_numbers)
```

```python
def remove_emojies(text):
    new=''
    for i in text:
        if i.isascii():
            new+=i
    return new
df['text']=df['text'].apply(remove_emojies)
```

```python
X_train=df['text']
y_train=df['emotion']

test_df=pd.read_csv('test.txt',sep=';',names=['text','emotion'], header=None)
X_test=test_df['text']
y_test=test_df['emotion'].map(emotion_numbers)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
bow_vectorizer=CountVectorizer()
X_train_bow=bow_vectorizer.fit_transform(X_train)
X_test_bow=bow_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix
model=MultinomialNB()
model.fit(X_train_bow,y_train)
y_pred=model.predict(X_test_bow)
accuracy=model.score(X_test_bow,y_test)
print('accuracy:',accuracy)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
tfidf_vectorizer=TfidfVectorizer()
X_train_tfidf=tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf=tfidf_vectorizer.transform(X_test)
```

```python
model_tfidf=MultinomialNB()
model_tfidf.fit(X_train_tfidf,y_train)
y_pred=model_tfidf.predict(X_test_tfidf)
accuracy=model_tfidf.score(X_test_tfidf,y_test)
print('accuracy:',accuracy)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_lr=LogisticRegression()
model_lr.fit(X_train_bow,y_train)
y_pred_lr=model_lr.predict(X_test_bow)
accuracy_lr=model_lr.score(X_test_bow,y_test)   
print('accuracy:',accuracy_lr)
```

```python
model_lr=LogisticRegression()
model_lr.fit(X_train_tfidf,y_train)
y_pred_lr=model_lr.predict(X_test_tfidf)
accuracy_lr=model_lr.score(X_test_tfidf,y_test)   
print('accuracy:',accuracy_lr)
```
